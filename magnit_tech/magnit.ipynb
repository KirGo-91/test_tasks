## Задание 2.

Реализовать алгоритм, который определит, является ли строка палиндромом.  
Тестовые примеры:  
'qwerewq' - палиндром  
'qwerrewq' - палиндром  
'adssdfgf' - не палиндром  
'r' - палиндром  
'rr' - палиндром  

```python
def is_palindrome(string):
  clear_string = string.lower().replace(' ', '')
  l = len(clear_string) // 2
  c = 0

  for i in range(l):
    if clear_string[i] == clear_string[-i-1]:
      c += 1

  if l == c:
    print(string, ' - палиндром')
  else:
    print(string, ' - не палиндром')
```

```python
is_palindrome('Нажал кабан на баклажан')
```

```python
def is_palindrome_2(string):
  clear_string = string.lower().replace(' ', '')
  reverse_string = clear_string[::-1]

  if clear_string == reverse_string:
    print(string, ' - палиндром')
  else:
    print(string, ' - не палиндром')
```

```python
is_palindrome_2('А роза упала на лапу Азора')
```
